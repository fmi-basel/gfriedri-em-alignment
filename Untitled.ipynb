```python
for f in file_list[0]:
    f = f.replace("\\", "/")
    target_dir = join(target_root, dirname(f))
    makedirs(target_dir, exist_ok=True)
    if not exists(join(target_dir, basename(f))):
        copyfile(join(root_dir, f), join(target_dir, basename(f)))
```
